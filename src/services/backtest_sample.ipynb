```python
from analisys.backtest import Backtest
from skopt import forest_minimize
```

```python
''' 
Algumas considerações sobre essa parte do codigo:
-preciso voltar com a otimização do long spread e short spread
-preciso normalizar os valores pra ficar entre 0 e 1 no space: Posso fazer uma função de normalização e desnormalização

'''

class BacktestOptimizer:

    def __init__(self) -> None:
        
        self.backtest_class = Backtest()


    def __call__(self,
                 symbol_x,
                 symbol_y,
                 df):
        
        space = [(1.0, 3.0),   #short_spread
                 (-3.0, -1.0), #long_spread
                (5, 20),         #period
                (0.01, 0.05),     #tp
                (-0.05, -0.01)]   #sl
        
        def simple_backtest(args):
            
            short_spread, long_spread, period, take_profit, stop_loss = args
            run_backtest = self.backtest_class('BOVA11', 'DOL$', df, 'close', short_spread, 
                                                    long_spread, 10000, period, take_profit, stop_loss, 'roi')
            
            return -run_backtest

        run_optimizer = forest_minimize(simple_backtest, space, random_state=1, verbose=1, n_calls = 50)

        return run_optimizer
```

```python
backtest_optimizer = BacktestOptimizer()
run = backtest_optimizer('BOVA11', 'DOL$', hist_list)
```
